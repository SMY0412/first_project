```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = np.linspace(0, 10, 10000)
```

```python
x
```

```python
y = np.sin(x) + 2
```

```python
y
```

```python
plt.plot(x, y)
```

```python
plt.savefig("sin.png", dpi = 100)
plt.savefig("sin.pdf", dpi = 100)
```

```python
fig, ax = plt.subplots()
ax.plot(x, y)
ax.grid(True)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Sin')
plt.show()
```

```python
norm = np.random.randn(10000) * 5 + 10
fig, ax = plt.subplots()
ax.hist(norm, bins=20, edgecolor='black')
ax.grid(True)
ax.set_xlabel('Value')
ax.set_ylabel('Frequency')
ax.set_title('Hist')
plt.savefig("hist.pdf", dpi = 100)
plt.savefig("hist.png", dpi = 100)
```

```python
sizes = [10, 20, 15, 5]
labels = ["sing", "dance", "rap", "basketball"]
explodes = [0, 0, 0.1, 0]
fig, ax = plt.subplots()
plt.pie(sizes, explode=explodes, labels=labels, startangle=90)
plt.title('Pie')
plt.axis('equal')
plt.savefig("pie.pdf", dpi = 50)
plt.savefig("pie.png", dpi = 50)
```
