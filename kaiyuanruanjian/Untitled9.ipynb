```python
import libcst
```

```python
class IntegerInFunc(libcst.CSTVisitor):
    def __init__(self):
        self.current_function = None
        self.function_integer = dict()

    def visit_FunctionDef(self, node):
        self.current_function = node.name.value
        self.function_integer[self.current_function] = list()

    def leave_FunctionDef(self, node):
        self.current_function = None

    def visit_Integer(self, node):
        if self.current_function:
            value = node.value
            self.function_integer[self.current_function].append(value)
```

```python
text = """
def foo():
    bar = 10
    return bar + 20
def buz():
    return "hello"
foo(30)
"""
```

```python
code = libcst.parse_module(text)
visitor = IntegerInFunc()
code.visit(visitor)
print(visitor.function_integer)
```
